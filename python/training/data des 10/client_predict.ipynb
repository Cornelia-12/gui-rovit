```python
import joblib
import pandas as pd
import heartpy as hp
import numpy as np

data = {"age": 45.0, "weight": 62.0, 'ppg': [int(float(e)) for e in list(pd.read_csv('client_ppg_data.csv').columns)]}
# data = {"age": 45.0, "weight": 62.0, 'ppg': [123,123,123]}
print(data) 

wd, m = hp.process(np.array(data['ppg']), sample_rate = 100.0)

data.pop("ppg")
    
data = pd.DataFrame([{**data, **m}])

scaler = joblib.load('scaler.pkl')

data = scaler.transform(data)

systole_model = joblib.load('systole_model.pkl')
diastole_model = joblib.load('diastole_model.pkl')

estimated_systole = systole_model.predict(data)
estimated_diastole = diastole_model.predict(data)

print(estimated_systole)
print(estimated_diastole)
```
